```python
def get_authors(i):  # extends all authors into a single list 
    #print 'Authors\n'
    authors=[]
    if 'author' in  i[1].keys(): 
        for j in i[1]['author'].split(' and '):        
            #print j 
            authors.append(j)
    return(authors)

Alist=[]
for i in bibtex_database.entries_dict.items():
    #print i[1]['author']
    Alist.extend(get_authors(i))
```

```python
def find_author_count(Alist,author='Kevin'):  ## uses the single list to count 1 person
    manual_count=0
    for i in Alist:
        if i.find(author)>-1:
            manual_count+=1
    return(manual_count)

def find_author_count2(items1,author1='Kevin', author2='Jayne'): ### uses the bibtex extry dict to count two authors
    count2=0
    for i in items1:
        #print i[1]['author']
        #Alist.extend(print_authors(i))
        authors=get_authors(i)
        if (find_author_count(authors,author1)>0) & (find_author_count(authors,author2)>0):
            count2+=1
    return(count2)



print 'Kevin\t\t',find_author_count(Alist,'Kevin')
#print find_author_count(Alist,'Provenzano')
print 'Skala-Walsh\t',find_author_count2(bibtex_database.entries_dict.items(),'Skala','Walsh')
print 'Enrico-Michelle\t',find_author_count2(bibtex_database.entries_dict.items(),'Enrico','Michelle')
```
